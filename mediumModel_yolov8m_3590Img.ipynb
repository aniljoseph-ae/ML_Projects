## Hyper paarameter tuning with Optuna

```python
%pip install optuna ultralytics

```

```python
# Install dependencies
!pip install optuna ultralytics

import optuna
from ultralytics import YOLO

def objective(trial):
    # Define hyperparameters to tune
    lr0 = trial.suggest_loguniform('lr0', 1e-6, 1e-2)
    batch_size = trial.suggest_categorical('batch_size', [8, 16, 32])
    imgsz = trial.suggest_categorical('imgsz', [640, 800, 1024])

    # Train the YOLOv8 model with the selected hyperparameters
    vision_model = YOLO('yolov8m.pt')

    # Use Optuna-suggested parameters for training
    vision_model.train(
        data='/content/drive/MyDrive/aircraft_reporting_project/mediumModel_yoloV8m/input/data.yaml',  # Dataset YAML file
        epochs=50,                           # Number of epochs
        imgsz=imgsz,                         # Image size
        batch=batch_size,                    # Batch size
        device=0,                            # Use GPU
        optimizer='Adam',                    # Optimizer
        lr0=lr0,                             # Initial learning rate
        weight_decay=0.0005,                 # Regularization
        momentum=0.937,                      # Momentum
        project='/content/drive/MyDrive/"aircraft_reporting_project"/mediumModel_yoloV8m/output_optimized',  # Save project folder
        name=f'trial_{trial.number}',        # Unique name for each trial
        workers=4,                           # Number of workers
        half=True,                           # Use FP16 precision
    )

    # Retrieve validation loss (or any metric Optuna optimizes)
    results_path = f'/content/drive/MyDrive/"aircraft_reporting_project"/mediumModel_yoloV8m/output_optimized/trial_{trial.number}/results.csv'
    try:
        # Read the results.csv file to get the validation loss
        with open(results_path, 'r') as f:
            lines = f.readlines()
            validation_loss = float(lines[-1].split(',')[-1])  # Assuming loss is the last column
    except:
        validation_loss = float('inf')  # Set to infinity if results are not available

    return validation_loss

# Set up and run Optuna optimization
study = optuna.create_study(direction='minimize')
study.optimize(objective, n_trials=10)

# Extract the best trial
best_trial = study.best_trial
print(f"Best Trial: {best_trial.number}")
print(f"Best Hyperparameters: {best_trial.params}")

# Re-train the model with the best hyperparameters
best_params = best_trial.params
vision_model = YOLO('yolov8m.pt')

vision_model.train(
    data='/content/drive/MyDrive/aircraft_reporting_project/mediumModel_yoloV8m/input/data.yaml',
    epochs=50,
    imgsz=best_params['imgsz'],
    batch=best_params['batch_size'],
    device=0,
    optimizer='Adam',
    lr0=best_params['lr0'],
    weight_decay=0.0005,
    momentum=0.937,
    project='/content/drive/MyDrive/"aircraft_reporting_project"/mediumModel_yoloV8m/output_optimized',
    name='mediumModel_yolov8m_3598img_optimized',  # Name for the best model
    workers=4,
    half=True,
)

# Save the best model explicitly
vision_model.save('/content/drive/MyDrive/"aircraft_reporting_project"/mediumModel_yoloV8m/output_optimized/best_modelM.pt')
```
